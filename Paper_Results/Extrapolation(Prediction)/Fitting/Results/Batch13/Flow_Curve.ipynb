```python
Y_A = np.genfromtxt('Y_A.txt')
Y_M_PL = np.genfromtxt('Y_M_PL.txt')
Y_M_HB = np.genfromtxt('Y_M_HB.txt')
#Y_M_CA = np.genfromtxt('Y_M_CA.txt')
Y_M_HBC = np.genfromtxt('Y_M_HBC.txt')
Y_D = np.genfromtxt('Y_D.txt')
ShR = np.genfromtxt('ShR.txt')

Y_A = Y_A.reshape((-1,1))
Y_M_PL = Y_M_PL.reshape((-1,1))
Y_M_HB = Y_M_HB.reshape((-1,1))
#Y_M_CA = Y_M_CA.reshape((-1,1))
Y_M_HBC = Y_M_HBC.reshape((-1,1))
Y_D = Y_D.reshape((-1,1))
ShR = ShR.reshape((-1,1))
```
