```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))
HTML("""
<style>
.output_png {
    display: table-cell;
    text-align: center;
    vertical-align: middle;
}
</style>
""")
```

```python
Y_A_03M = np.genfromtxt('Y_A_03M.txt')
Y_M_03M = np.genfromtxt('Y_M_03M.txt')
Y_D_03M = np.genfromtxt('Y_D_03M.txt')
ShR_03M = np.genfromtxt('ShR_03M.txt')
Y_A_06M = np.genfromtxt('Y_A_06M.txt')
Y_M_06M = np.genfromtxt('Y_M_06M.txt')
Y_D_06M = np.genfromtxt('Y_D_06M.txt')
ShR_06M = np.genfromtxt('ShR_06M.txt')
Y_A_09M = np.genfromtxt('Y_A_09M.txt')
Y_M_09M = np.genfromtxt('Y_M_09M.txt')
Y_D_09M = np.genfromtxt('Y_D_09M.txt')
ShR_09M = np.genfromtxt('ShR_09M.txt')
Y_A_12M = np.genfromtxt('Y_A_12M.txt')
Y_M_12M = np.genfromtxt('Y_M_12M.txt')
Y_D_12M = np.genfromtxt('Y_D_12M.txt')
ShR_12M = np.genfromtxt('ShR_12M.txt')

Y_A_03M = Y_A_03M.reshape((-1,1))
Y_M_03M = Y_M_03M.reshape((-1,1))
Y_D_03M = Y_D_03M.reshape((-1,1))
ShR_03M = ShR_03M.reshape((-1,1))
Y_A_06M = Y_A_06M.reshape((-1,1))
Y_M_06M = Y_M_06M.reshape((-1,1))
Y_D_06M = Y_D_06M.reshape((-1,1))
ShR_06M = ShR_06M.reshape((-1,1))
Y_A_09M = Y_A_09M.reshape((-1,1))
Y_M_09M = Y_M_09M.reshape((-1,1))
Y_D_09M = Y_D_09M.reshape((-1,1))
ShR_09M = ShR_09M.reshape((-1,1))
Y_A_12M = Y_A_12M.reshape((-1,1))
Y_M_12M = Y_M_12M.reshape((-1,1))
Y_D_12M = Y_D_12M.reshape((-1,1))
ShR_12M = ShR_12M.reshape((-1,1))
```

# Flow Curves

## 3rd Month

```python
import matplotlib.font_manager
from matplotlib import rc
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)

plt.figure(figsize=(10,10),dpi=300)

plt.rc('font', size=20)          # controls default text sizes
plt.rc('axes', titlesize=20)     # fontsize of the axes title
plt.rc('axes', labelsize=22)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
plt.rc('ytick', labelsize=20)    # fontsize of the tick labels
plt.rc('legend', fontsize=22)    # legend fontsize
plt.rc('figure', titlesize=22)  # fontsize of the figure title

plt.plot(np.power(10,ShR_03M[0::2]),np.power(10,Y_A_06M[0::2]) , 'bP-', markerfacecolor='none', linewidth=1.5, markersize = 11.0)
plt.plot(np.power(10,ShR_03M[1::2]),np.power(10,Y_D_06M[1::2]) , 'rd-', markerfacecolor='none', linewidth=1.5, markersize = 11.0)
plt.plot(np.power(10,ShR_03M[0::2]),np.power(10,Y_M_06M[0::2]) , 'gs-' , markerfacecolor='none', linewidth=1.5, markersize = 11.0)

plt.xlabel('Shear Rate [1/s]', fontname = 'Times New Roman', fontsize=40)
plt.ylabel('Viscosity [Pa.s]', fontname = 'Times New Roman', fontsize=40)
plt.legend(['Experiment','DNN','MNN'],loc='upper right')
#plt.title('Actual and Predicted Flow Curve for a Random Batch')

plt.xticks(np.arange(0,10, 1.0),fontname = "Times New Roman", fontsize=30)
plt.yticks(fontname = "Times New Roman", fontsize=30)
#plt.ylim(0.3,3000)
plt.xscale("log")
plt.yscale("log")

plt.savefig('PredictionUnknownAgingB18M03.png',bbox_inches = 'tight')

plt.show()
```
