```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))
HTML("""
<style>
.output_png {
    display: table-cell;
    text-align: center;
    vertical-align: middle;
}
</style>
""")
```

```python
Y_A = np.genfromtxt('UpTp02Batch13Y_A.txt')
Y_M02 = np.genfromtxt('UpTp02Batch13Y_M.txt')
Y_D02 = np.genfromtxt('UpTp02Batch13Y_D.txt')
Y_M05 = np.genfromtxt('UpTp05Batch13Y_M.txt')
Y_D05 = np.genfromtxt('UpTp05Batch13Y_D.txt')
Y_M08 = np.genfromtxt('UpTp08Batch13Y_M.txt')
Y_D08 = np.genfromtxt('UpTp08Batch13Y_D.txt')
Y_M11 = np.genfromtxt('UpTp11Batch13Y_M.txt')
Y_D11 = np.genfromtxt('UpTp11Batch13Y_D.txt')
ShR = np.genfromtxt('UpTp02Batch13ShR.txt')

Y_A = Y_A.reshape((-1,1))
Y_M02 = Y_M02.reshape((-1,1))
Y_D02 = Y_D02.reshape((-1,1))
Y_M05 = Y_M05.reshape((-1,1))
Y_D05 = Y_D05.reshape((-1,1))
Y_M08 = Y_M08.reshape((-1,1))
Y_D08 = Y_D08.reshape((-1,1))
Y_M11 = Y_M11.reshape((-1,1))
Y_D11 = Y_D11.reshape((-1,1))
ShR = ShR.reshape((-1,1))
```

# Flow Curves

```python
import matplotlib.font_manager
from matplotlib import rc
rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)

plt.figure(figsize=(10,10),dpi=300)

plt.rc('font', size=20)          # controls default text sizes
plt.rc('axes', titlesize=20)     # fontsize of the axes title
plt.rc('axes', labelsize=22)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=20)    # fontsize of the tick labels
plt.rc('ytick', labelsize=20)    # fontsize of the tick labels
plt.rc('legend', fontsize=22)    # legend fontsize
plt.rc('figure', titlesize=22)  # fontsize of the figure title

plt.plot(np.power(10,ShR),np.power(10,Y_A) , 'k-', markerfacecolor='none', linewidth=3, markersize = 11.0)
plt.plot(np.power(10,ShR[0::2]),np.power(10,Y_M02[0::2]) , 'r<:' , markerfacecolor='none', linewidth=1.5, markersize = 11.0)
plt.plot(np.power(10,ShR[1::3]),np.power(10,Y_M05[1::3]) , 'yo:', markerfacecolor='none', linewidth=1.5, markersize = 11.0)
plt.plot(np.power(10,ShR[0::3]),np.power(10,Y_M08[0::3]) , 'bd:', markerfacecolor='none', linewidth=1.5, markersize = 11.0)
plt.plot(np.power(10,ShR[2::3]),np.power(10,Y_M11[2::3]) , 'gP:', markerfacecolor='none', linewidth=1.5, markersize = 11.0)

plt.xlabel('Shear Rate [1/s]', fontname = 'Times New Roman', fontsize=40)
plt.ylabel('Viscosity [Pa.s]', fontname = 'Times New Roman', fontsize=40)
plt.legend(['Experiment','Up to 2 Months','Up to 5 Months','Up to 8 Months','Up to 11 Months'],loc='upper right')
#plt.title('Actual and Predicted Flow Curve for a Random Batch')

plt.xticks(np.arange(0,10, 1.0),fontname = "Times New Roman", fontsize=30)
plt.yticks(fontname = "Times New Roman", fontsize=30)
#plt.ylim(0.6,13)
plt.xscale("log")
plt.yscale("log")

plt.savefig('AginWithinBatch18MFNN.png',bbox_inches = 'tight')

plt.show()
```
